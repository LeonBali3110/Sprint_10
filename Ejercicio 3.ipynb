```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from time import sleep
```

# Ejercicio 3
La documentación de este dataset esta en el PDF adjunto en el Github

```python
driver = webdriver.Chrome()
driver.get("https://www.warhammer.com/es-ES/shop/warhammer-40000/space-marines")
sleep(5)
```

Llamo a la página, y le doy una espara de 5 segundos para darle tiempo de cargar a la página.

```python
cookies = driver.find_element(by=By.ID, value="onetrust-accept-btn-handler")
```

```python
cookies.click()
```

```python
close = driver.find_element(by=By.XPATH, value='//*[@id="__next"]/div/div/section[1]/div/section/div/div/button')
close.click()
```

Le da click a aceptar las cookies y que cierre el selector de idioma (ya en la URL lo ponemos en Español)

```python
submit_button = driver.find_element(by=By.ID, value="show-more")
```

```python
for _ in range(6):
    submit_button.click()
    sleep(5)
```

Para sacar la mayor cantidad de datos le doy al botón de 'mostrar más'. Lo configuro para que lo presione 6 veces ya que esa es la cantidad de veces necesaria para tener todos los productos.

Le doy un tiempo de espera de 5 segundos para que tenga tiempo de cargar el nuevo contenido.

```python
Productos = driver.find_elements(By.XPATH, '//*[@id="__next"]/div/div/div/section/div/div/div/div/div/div/div[5]/div/div/div[2]/div/a/div/span')
Precios = driver.find_elements(By.XPATH, '//*[@id="__next"]/div/div/div/section/div/div/div/div/div/div/div[5]/div/div/div[2]/div/div[2]/div[1]/span')
```

```python
df_data = {'Producto':[], 'Precio':[]}
```

```python
for Producto, Precio in zip(Productos, Precios):
    print("Producto:", Producto.text)
    print("Precio:", Precio.text)
    print()
    df_data['Producto'].append(Producto.text)
    df_data['Precio'].append(Precio.text)
```

Busco los datos y aprovecho para colocarlos en un diccionario.

```python
df = pd.DataFrame(df_data)
```

Finalmente aquí esta el dataset.

```python
driver.quit()
```
