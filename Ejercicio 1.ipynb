# Frases (Beutiful Soup)

Primero hice un dataset de frases usando la web 'Quotes to Scrape' con sus autores y utilize el metodo de Beutiful Soup.

```python
URL = "https://quotes.toscrape.com/"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
quotes = soup.find_all("div", class_="quote")
```

Pedi la URL, carge el html y busque las cajas donde se encuentra la información.

```python
df_data = {'Frase':[], 'Autor':[]}
```

```python
for quote in quotes:
    frase = quote.find("span", itemprop="text")
    autor = quote.find("small", itemprop="author")
    df_data['Frase'].append(frase.text)
    df_data['Autor'].append(autor.text)
    print(frase.text)
    print(autor.text)
    print()
```

Separé la frases y los autores del resto de la información y lo guardé en un diccionario.

```python
df = pd.DataFrame(df_data)
```

```python
df
```

Convertí el diccionario en un dataset.

# Frases (Selenium)

Esto es lo mismo, pero usando Selenium.

```python
driver = webdriver.Chrome()
driver.get("https://quotes.toscrape.com/")
driver.title
```

Abrí el navegador Chrome y la página de Quotes.

```python
df_data_2 = {'Frase':[], 'Autor':[]}
```

```python
messages = driver.find_elements(By.CLASS_NAME, "text")
persons = driver.find_elements(By.CLASS_NAME, "author")
```

Busque el lugar donde encontre las frases y autores.

```python
for message, person in zip(messages, persons):
    print("Frase:", message.text)
    print("Autor:", person.text)
    print()
    df_data_2['Frase'].append(message.text)
    df_data_2['Autor'].append(person.text)
```

```python
df_2 = pd.DataFrame(df_data_2)
```

# BME (Selenium)

```python
driver2 = webdriver.Chrome()
driver2.get(URL2)
driver2.title
```

```python
col_sel = driver2.find_elements(By.XPATH, '//*[@id="root"]/div[3]/table/thead/tr/th')
```

```python
columnas_df = []
```

```python
for columna in col_sel:
    print(columna.text)
    columnas_df.append(columna.text)
    print()
```

Lo primero que hice, en parte para hacer pruebas y familiarizarme con la web, es separar las columnas y guardarlas en una lista para más tarde usarlas para el dataframe.

```python
datos = driver2.find_elements(By.XPATH, '//*[@id="root"]/div[3]/table/tbody/tr/td')
```

```python
filas = []
```

```python
import re

for dato in datos:
    dato_text = dato.text.strip()  # Eliminar espacios en blanco alrededor del texto
    if dato_text.startswith('Suspendido'):
        fecha_match = re.search(r'\d{2}/\d{2}/\d{4}', dato_text)  # Buscar la fecha con una expresión regular
        if fecha_match:
            fecha_suspendido = fecha_match.group()
            filas.append('Suspendido')
            print('Suspendido')
            filas.append(fecha_suspendido)
            print(fecha_suspendido)
        else:
            filas.append(dato_text)# Si no se encuentra la fecha, agregar el texto completo
            print(dato.text)
    else:
        filas.append(dato_text)
        print(dato.text)
    print()
```

Aquí busque los datos de la tabla, pero me encontre con que dos de las filas habian fucionado dos sus celdas por estar suspendidas, lo que termino dandome muchos problemas al momento organizar lo datos. Así tambien aproveche para separar

```python
datos_df = []
```

```python
for i in range(0, len(filas), 9):
    grupo = filas[i:i+9]
    datos_df.append(grupo)
```

```python
datos_df
```

```python
df_BME = pd.DataFrame(datos_df, columns = columnas_df)
```
